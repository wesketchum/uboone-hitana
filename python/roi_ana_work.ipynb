```python
import ROOT
ROOT.gStyle.SetOptStat(0)
ROOT.ErrorIgnoreLevel=ROOT.kWarning
ROOT.enableJSVis()
```

```python
#globals
mycanvas = ROOT.TCanvas("mycanvas","MyCanvas")
files = {}
compare_trees = {}
h = {}
```

```python
#functions

def add_roi_file(fname,marker,color=ROOT.kBlack):
    files[marker] = ROOT.TFile(fname,"READ")
    compare_trees[marker] = files[marker].Get("tree_roi")
    compare_trees[marker].SetLineColor(color)

def create_hist(hname,title,nbins,xlow,xhi,color=ROOT.kBlack,width=2):
    h[hname] = ROOT.TH1F(hname,title,nbins,xlow,xhi)
    h[hname].SetLineColor(color)
    h[hname].SetLineWidth(width)
    return

def create_compare_hists(hname_base,var_name,title,nbins,xlow,xhi,var_cut="1",plot_opts=""):
    for name, tree in compare_trees.iteritems():
        hist_name = "h_%s_%s"%(name,hname_base)
        create_hist(hist_name,title,nbins,xlow,xhi,tree.GetLineColor())
        tree.Project(hist_name,var_name,var_cut,plot_opts)
        print "Created histogram %s" % hist_name
    return

def draw_collection(hname_base,norm=False):
    first=True
    global leg
    leg = ROOT.TLegend(0.7,0.7,0.9,0.9)
    for name in compare_trees:
        hist_name = "h_%s_%s"%(name,hname_base)
        if first:
            if norm:
                h[hist_name].DrawNormalized()
            else:
                h[hist_name].Draw()
            first=False
        else:
            if norm:
                h[hist_name].DrawNormalized("same")
            else:
                h[hist_name].Draw("same")
        leg.AddEntry(hist_name,name,"l")
    leg.Draw()
    return
```

```python
add_roi_file("/uboone/data/users/wketchum/roiana_mcc8_MarcoFilter.root","mcc8",ROOT.kRed)
add_roi_file("/uboone/data/users/wketchum/roiana_mcc7_MarcoFilter.root","mcc7",ROOT.kBlue)
add_roi_file("/uboone/data/users/wketchum/roiana_mcc8_MarcoFilter_ROI.root","mcc8_roi",ROOT.kBlack)
add_roi_file("/uboone/data/users/wketchum/roiana_mcc7_MarcoFilter_315.root","mcc7_3_15",ROOT.kGreen+2)
add_roi_file("/uboone/data/users/wketchum/roiana_mcc7_MarcoFilter_41.root","mcc7_4_1",ROOT.kViolet)
```

```python
create_compare_hists("baseline","integral_last20/20","Baseline in last 20 ticks;adc;N_{ROIs}",100,-40,60)
ROOT.enableJSVis()
draw_collection("baseline",True)
mycanvas.SetLogy()
mycanvas.Draw()
```
